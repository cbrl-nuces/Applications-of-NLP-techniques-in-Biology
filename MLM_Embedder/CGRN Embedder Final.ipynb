```python
class NonZeroCCE(tf.keras.losses.Loss):
    def __init__(self,Weights=[],smoothing = None):
        super().__init__()
        if smoothing is not None:
            self.cce = tf.keras.losses.CategoricalCrossentropy(reduction='none', label_smoothing=smoothing)
        else:
            self.cce = tf.keras.losses.CategoricalCrossentropy(reduction='none')
        self.weights = np.array(Weights)
    def call(self, y_true, y_pred):
        if len(self.weights)>0:
            loss = self.cce(self.weights*y_true,y_pred)
        else:
            loss = self.cce(y_true,y_pred)
        return tf.math.reduce_sum(loss)/((y_true.shape[0]*y_true.shape[1])-float(tf.math.reduce_sum(tf.cast(tf.equal(tf.math.reduce_sum(y_true,-1),0),tf.int32))))

def Accuracy(y_true,y_pred):
    y_true = y_true.numpy()
    y_pred = y_pred.numpy()

    m = np.sum(y_true, axis=-1)
    y_t = np.argmax(y_true[m==1],axis=-1)
    y_p = np.argmax(y_pred[m==1],axis=-1)
    return accuracy_score(y_t,y_p)

def dump_pickle(path,files):
    with open(path, 'wb') as f:
        pickle.dump(files, f)
def load_pickle(path):
    with open(path, 'rb') as f:
        files = pickle.load(f)
    return files

def plot_hist(files):    
    lens = []
    for idx in tqdm(range(len(files))):
        path = files[idx]
        protein_sequence = np.load(path)
        lens.append(len(protein_sequence))
    plt.hist(lens)
    plt.show()
```

```python
class MLM_Loader(tfk.utils.Sequence):
    def __init__(self, files, batch_size=30, shuffle=True, masked_perc=0.0, max_seq_len=2000, train=False,loss="MASKED"):
        self.files = files
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.masked_perc = masked_perc
        self.loss = loss
        self.max_seq_len = max_seq_len
        self.mask = embedding_keys['<mask>']
        self.length = len(files)
        self.indices = np.arange(0,self.length)
        self.train = train
        self.on_epoch_end()
        
    def generateBatches(self):
        self.files_ = self.files[self.indices]
    
    def mask_data(self, protein_sequence, length):
        temp_label = copy.deepcopy(protein_sequence)
        choices = np.random.choice(3, length, p=[1-self.masked_perc,self.masked_perc*0.8,self.masked_perc*0.2])
        if True in (choices==1) or True in (choices==2):
            protein_sequence[:length][choices==1] = self.mask
            protein_sequence[:length][choices==2] = np.random.choice(25, np.sum(choices==2))+1
            if self.loss=="MASKED":
                temp_label[:length][choices==0] = 0
        return (protein_sequence,temp_label)
    
    #Convert Amino Acids to their IDs and pad with <pad> token if size<max_seq_len
    def preprocess(self,protein_sequence):
        length = len(protein_sequence)
        protein_sequence = np.concatenate((protein_sequence, [0]*(self.max_seq_len-len(protein_sequence))), axis=0)
        if self.masked_perc>0.0:
            protein_sequence,true_label = self.mask_data(protein_sequence,length)
        else:
            true_label = copy.deepcopy(protein_sequence)
        return np.array([protein_sequence,true_label])

    def fetch_data(self, path):
        return self.preprocess(np.load(path))

    def on_epoch_end(self):
        seed_value = np.random.randint(1,100)
        #print("Seed Value:",seed_value)
        np.random.seed(seed_value)
        if self.shuffle:
            np.random.shuffle(self.indices)
        self.generateBatches()
        
    def __getitem__(self,index):
        data = np.array([self.fetch_data(file) for file in self.files_[index*self.batch_size:(index+1)*self.batch_size]])
        prot_seq = data[:,0,:]
        labels = data[:,1,:]
        labels = np.array(tf.one_hot(labels,len(embedding_keys))[:,:,1:len(embedding_keys)-1])
        prot_seq = np.array(prot_seq)
        labels = np.array(labels)
        if self.train:
            return ([
                    prot_seq
                   ],
                    labels
                   )
        else:
            return [prot_seq]

    def __len__(self):
        if self.length%self.batch_size!=0:
            return (self.length//self.batch_size)+1
        return self.length//self.batch_size
```

```python
with strategy.scope():
    input1_ = tfk.layers.Input(shape=(2000,), name='sequence_input')
    
    emb = tfk.layers.Embedding(input_dim=len(embedding_keys), output_dim=1000, name='embds')(input1_)
    #Embedder
    x1 = tfk.layers.Conv1D( 100, 11, strides=1, padding='same', activation='relu')(emb)
    x2 = tfk.layers.Conv1D( 100, 7, strides=1, padding='same', activation='relu')(emb)
    x3 = tfk.layers.Conv1D( 100, 3, strides=1, padding='same', activation='relu')(emb)
    x = tfk.layers.concatenate([x1, x2, x3], axis=-1)
    
    x1 = tfk.layers.Conv1D( 100, 11, strides=1, padding='same', activation='relu')(x)
    x2 = tfk.layers.Conv1D( 100, 7, strides=1, padding='same', activation='relu')(x)
    x3 = tfk.layers.Conv1D( 100, 3, strides=1, padding='same', activation='relu')(x)
    x = tfk.layers.concatenate([x1, x2, x3], axis=-1)
    
    x1 = tfk.layers.Conv1D( 100, 11, strides=1, padding='same', activation='relu')(x)
    x2 = tfk.layers.Conv1D( 100, 7, strides=1, padding='same', activation='relu')(x)
    x3 = tfk.layers.Conv1D( 100, 3, strides=1, padding='same', activation='relu')(x)
    x = tfk.layers.concatenate([x1, x2, x3], axis=-1)

    gru_out = tfk.layers.Bidirectional(tfk.layers.GRU(units=150, return_sequences=True), name='bigru1')(emb)
    gru_out2 = tfk.layers.Bidirectional(tfk.layers.GRU(units=150, return_sequences=True), name='bigru2')(gru_out)
    x = tfk.layers.concatenate([gru_out2, x], axis=-1)
    
    x = tfk.layers.TimeDistributed( tfk.layers.Dense(500, activation='relu'), name='output1')(x)
    x = tfk.layers.TimeDistributed( tfk.layers.Dense(500, activation='relu'), name='output2')(x)
    output_ = tfk.layers.TimeDistributed( tfk.layers.Dense(len(embedding_keys)-2, activation='softmax') ,name='output')(x)
    model = tfk.models.Model([input1_], output_)
    model.compile(loss=NonZeroCCE(), metrics=[Accuracy], optimizer='adam', run_eagerly = True)
    
    model.summary()
```
